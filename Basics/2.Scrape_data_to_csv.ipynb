# **Taking List of the largest companies (wikipedia)**

```python
from bs4 import BeautifulSoup
import requests
```

```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_India"
```

```python
# Define headers FIRST
user_agent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
headers = {'User-Agent': user_agent}

# Then pass headers to requests.get
page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, 'html.parser')
```

# **The first table**

```python
table = soup.find_all('table')[0]

world_titles = table.find_all('th')

world_table_title = [title.text.strip() for title in world_titles]

df = pd.DataFrame(columns = world_table_title)
column_data = table.find_all('tr')

for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,12))
plt.bar(df['Name'][-1::-1], df['Revenue(billions US$)'][-1::-1])
plt.xlabel('Name')
plt.xticks(rotation=90)
plt.ylabel('Revenue(billions US$)')
plt.title('Revenue of Top 70 Indian Companies by Revenue')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(20,12))
plt.bar(df['Name'][-1::-1], df['Profit(billions US$)'][-1::-1])
plt.xlabel('Name')
plt.xticks(rotation=90)
plt.ylabel('Profit(billions US$)')
plt.title('Profit of Top 70 Indian Companies by Profit')
plt.grid(True)
plt.show()
```

```python
df['Profit(billions US$)'] = pd.to_numeric(df['Profit(billions US$)'], errors='coerce')
```

```python
df
```

```python
df['Profit(billions US$)'] = df['Profit(billions US$)'].fillna(0)
```

```python
plt.figure(figsize=(20,12))
plt.pie(df['Profit(billions US$)'][:20], labels=df['Name'][:20], autopct='%1.1f%%', startangle=140, explode=[0.1 if x < 1 else 0 for x in range(20)])
plt.title('Profit of Top 20 Indian Companies by Profit')
plt.show()
```

```python
df['Revenue(billions US$)'] = pd.to_numeric(df['Revenue(billions US$)'], errors='coerce')
df['Revenue(billions US$)'] = df['Revenue(billions US$)'].fillna(0)
plt.figure(figsize=(20,12))
plt.pie(df['Revenue(billions US$)'][:20], labels=df['Name'][:20], autopct='%1.1f%%', startangle=140, explode=[0.1 if x < 1 else 0 for x in range(20)])
plt.title('Revenue of Top 20 Indian Companies by Revenue')
plt.show()
```
